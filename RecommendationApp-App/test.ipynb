```python
filtered_df = df[["striker_name","kadamba_bowler_name","runs","extras","extras_type_name","bwl_event_type"]]
 
filtered_df = filtered_df.copy()

# Calculating total_conceeded runs
filtered_df['total_conceeded'] =   filtered_df.apply(lambda row: row['runs'] + row['extras'] if row['extras_type_name'] not in ['legbyes', 'byes', None] else row['runs'], axis=1)

grouped_data1 =  filtered_df.groupby(['striker_name', 'kadamba_bowler_name']).agg(total_runs=('total_conceeded', 'sum')).reset_index()

new_filtered_df = filtered_df[filtered_df['bwl_event_type'] == 'legal_ball']

grouped_data2 =  new_filtered_df.groupby(['striker_name', 'kadamba_bowler_name']).agg(balls_faced=('bwl_event_type', 'count')).reset_index()

final_data = pd.merge(grouped_data1, grouped_data2, on=['striker_name', 'kadamba_bowler_name'])
final_data = final_data[['striker_name', 'kadamba_bowler_name', 'total_runs', 'balls_faced']]

final_data['economy_rate'] = (final_data['total_runs'] / final_data['balls_faced']) * 6
```
